```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.ones((700, 1200, 3), dtype=np.uint8)*255

cv2.circle(img,(1100,100),50,(255,255,0),thickness=-1)

cv2.rectangle(img, (0,600), (1200, 700), (0, 255,120), thickness=-1)

cv2.rectangle(img, (300,300), (400, 600), (185, 122,87), thickness=-1)
cv2.circle(img,(350,300),100,(170,255,0),thickness=-1)

cv2.circle(img,(800,370),40,(0,0,255),thickness=3)
cv2.rectangle(img, (700, 300), (1000, 600), (0, 0, 255), thickness=5)
cv2.rectangle(img, (900, 450), (1000, 600), (0, 0, 255), thickness=5)
cv2.line(img,(700, 300),(850,200),(0, 0, 255), thickness=5)
cv2.line(img,(850,200),(1000,300),(0, 0, 255), thickness=5)

cv2.putText(img,"SepehrPouladi",(50,100),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,2,(0,0,0),thickness=4)

plt.imshow(img)


```
