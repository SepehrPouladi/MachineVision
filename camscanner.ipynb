Open and Display Image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image


img = cv2.imread("download.png")

img_rgb = img[:,:, ::-1]

plt.imshow(img_rgb)

```

Theresholding

```python
img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)

img_scan = cv2.adaptiveThreshold(img_gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,15,7)

plt.imshow(img_scan,cmap="gray")
```

Increase & Decrease Contrast

```python
matrix_low_contrast = np.ones(img_scan.shape) * 0.76
matrix_high_contast = np.ones(img_scan.shape) * 1.1

img_scan_lower  = np.uint8(cv2.multiply(np.float64(img_scan), matrix_low_contrast))
img_scan_higher = np.uint8(np.clip(cv2.multiply(np.float64(img_scan), matrix_high_contast), 0, 255))


plt.figure(figsize=[20,5])
plt.subplot(131); plt.imshow(img_scan,cmap="gray");       plt.title("Original")
plt.subplot(132); plt.imshow(img_scan_lower,cmap="gray"); plt.title("Lower Contrast")
plt.subplot(133); plt.imshow(img_scan_higher,cmap="gray");plt.title("Higher Contrast")
```

Brighter & Darker
- we use Higher Contrast image in here

```python
matrix = np.ones(img_scan_higher.shape, dtype="uint8") * 50

img_rgb_brighter = cv2.add(img_scan_higher,matrix)
img_rgb_darker = cv2.subtract(img_scan_higher,matrix)

plt.figure(figsize=[20,5])
plt.subplot(131); plt.imshow(img_scan_higher,cmap = "gray"); plt.title("Higher Contrast Image")
plt.subplot(132); plt.imshow(img_rgb_brighter, cmap = "gray"); plt.title("brighter")
plt.subplot(133); plt.imshow(img_rgb_darker, cmap = "gray"); plt.title("darker")
```

Save Image!

```python
cv2.imwrite("scanneddocs.png",img_rgb_darker)
Image("scanneddocs.png")
```
